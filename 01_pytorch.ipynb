Reshaping,stacking,squeezing and unsqeezing tensors

```python
x=torch.arange(1.,10.)
x,x.shape
```

```python
x_reshaped=x.reshape(1,9)
x_reshaped
```

```python
z=x.view(1,9)
z,z.shape
```

```python
#changing z changes x(because view of a tensor shares the same memory as the orginial input)
z[:,0]=5
z,x    
```

```python
#stack tensors on top of each other
x_stack=torch.stack([x,x,x,x])
x_stack
```

```python
#torch squeeze().removes all single dimensions from a target tensor
print(f"previous tensor:{x_reshaped}")
print(f"previous shape :{x_reshaped.shape}")
x_reshaped.shape
```

```python
x_sequeezes=x_reshaped.squeeze()
print(f"new tensor:{x_sequeezes}")
```

```python
#torch unsqueeze()-adds a single dimension to a target tensor at a specific dim
print(f"previous target:{x_sequeezes}")
print(f"previous shape :{x_sequeezes.shape}")

#add an extra dimension with unsqeeze
x_unsequeezed=x_sequeezes.unsqueeze(dim=0)
print(f"new tensor:{x_unsequeezed}")
print(f"new shape:{x_unsequeezed.shape}")
```

```python
#torch permute-rearranges  the dimensions of a target tensor in a specified order
x_original=torch.rand(size=(224,224,3))#[height,width,colour_channel]

#permute the orginial tensor to rearrange the axis (or dim) order
x_permute=x_original.permute(2,0,1)#shift axis 0->1,1->2,2->0

print(f"previous shape:{x_original.shape}")
print(f"new shape:{x_permute.shape}")#[colout_channel,height,width]
```
