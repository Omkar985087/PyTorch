Indexing(selecting data from tensors)

```python
#indexing with pytorch is similar to indexing with numpy
import torch
import pandas as pd 
import numpy as np
```

```python
x=torch.arange(1,10).reshape(1,3,3)
x,x.shape
```

```python
#let index on new tensor
x[0]
```

```python
#let's index on middle bracket
x[0,0]
x[0][0]
```

```python
#let's index on inner bracket
x[0][0][2]
```

```python
#you can also use ":" to select "all" of a target dimension
x[:,0]
x[:,1]
```

```python
#get all values of 0th and 1st dimensions but only index 1 of 2nd dimension
x[:,:,1]
```

```python
#get all values of the 0 dimesnions but only the 1 index value of 1st and 2nd dimension
x[:,1,1]
```

```python
#get index 0 of oth and 1st dimension and all values of 2nd dimension
x[0,0,:]
```

```python
#index on x to return 3,6,9
x[:,:,2]
```

Pytorch tensors and numpy

```python
#numpy array to tensor
array=np.arange(1.0,8.0)
tensor=torch.from_numpy(array)#warning:when converting from numpy->pytorch reflect numpy's default numpy datatype of float64 unless specified otherwise
array,tensor
```

```python
#change the value of array,what will this do to tensor?
array=array+1
array,tensor
```

```python
tensor=torch.ones(7)
numpy_tensor=tensor.numpy()
tensor,numpy_tensor
```

```python
numpy_tensor.dtype
```

```python
#change the tensor,what happens to 'numpy_tensor'?
tensor=tensor+1
tensor,numpy_tensor
```
